```julia
using Plots, LaTeXStrings, LinearAlgebra
pyplot()

x = range(0,20, length=2) 
c1(x) = (2 - x)/2  #x1 + 2x2 >= 2
z(x) = dot([1,1], primal_optimal) - x

primal_optimal = [1, 1/2]
primal_pt1 = [0,0]
primal_pt2 = [0,1]
primal_pt3 = [1,0.5]

plot(Shape([(1,3), (1,0.5), (2,0), (3,0)]), fill = 0.2, alpha = 0.2,
    xaxis = (L"x_1", (-0.06,3)),
    yaxis = (L"x_2", (-0.06,3)),
    aspect_ratio = :equal,
    bordercolor= "white",
    label = "",
    xticks = 0:3,
    yticks = 0:3,
    size = (250, 250))

hline!([0], color = :black, label = "")
vline!([0], color = :black, label = "")

plot!(x, c1.(x), color = :1, label = "Constraints")
vline!([1], color = :1, label = "") 
plot!(x, z.(x), line = :dash, color = :3, label = "Obj. function")

quiver!([0,0],[0,1], quiver=([0,1/2],[1/2,-1/4]), color = :2, label = "")

scatter!([primal_pt1[1]], [primal_pt1[2]], color = :2, label = "")
scatter!([primal_pt2[1]], [primal_pt2[2]], color = :2, label = "")
scatter!([primal_pt3[1]], [primal_pt3[2]], color = :2, label = "")

annotate!(primal_pt1[1] + 0.05, primal_pt1[2] + 0.15, ("A", 9, :left))
annotate!(primal_pt2[1] + 0.05, primal_pt2[2] + 0.15, ("B", 9, :left))
annotate!(primal_pt3[1] + 0.05, primal_pt3[2] + 0.15, ("C", 9, :left))

savefig("primal_plot.pdf")
```

```julia
dual_optimal = [1/2, 1/2]
dual_pt1 = [0,0]
dual_pt2 = [0.5,0]
dual_pt3 = [0.5,0.5]

p = range(0,20, length=2) 
c1(p) = 1 - p 
z(p) = dot([2,1], dual_optimal) - 2p

plot(Shape([(0,0), (0,1), (1/2,1/2), (1/2,0)]), fill = 0.2, alpha = 0.2,
    xaxis = (L"p_1", (-0.03,1.5)),
    yaxis = (L"p_2", (-0.03,1.5)),
    bordercolor= "white",
    aspect_ratio = :equal,
    label = "",
    xticks = 0:1/2:2,
    yticks = 0:1/2:2,
    size = (250, 250))

hline!([0], color = :black, label = "")
vline!([0], color = :black, label = "")

plot!(x, c1.(x), color = :1, label = "Constraints")
vline!([1/2], color = :1, label = "") 
plot!(x, z.(x), line = :dash, color = :3, label = "Obj. function")

quiver!([0,1/2],[0,0], quiver=([1/4,0],[0,1/4]), color = :2, label = "")

scatter!([dual_pt1[1]], [dual_pt1[2]], color = :2, label = "")
scatter!([dual_pt2[1]], [dual_pt2[2]], color = :2, label = "")
scatter!([dual_pt3[1]], [dual_pt3[2]], color = :2, label = "")

savefig("dual_plot.pdf")
```
