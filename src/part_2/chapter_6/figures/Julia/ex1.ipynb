```julia
#Ex. 1 - Lecture 6

using Plots; pyplot()
using LaTeXStrings
```

```julia
n=1000

blue = RGBA{Float64}(0.0,0.6056031611752248,0.978680117569607,1.0)

f(x) = -12x[2] + 4x[1]^2 + 4x[2]^2 + 4(x[1]*x[2]) 

x1 = range(-50,50,length=n);
x2 = range(-50,50,length=n);

z = [f([x1[i],x2[j]]) for j = 1:n, i = 1:n];

contour(x1, x2, z, 
    levels = [-11.5, -9, -5, 2, 10, 20, 40],
    xaxis = (L"$x_1$", (-2,2)),
    yaxis = (L"$x_2$", (-1,3)),
    clims = (-20, 40),
    cbar = false,
    aspect_ratio = :equal,
    contour_labels = true,
    legend = true,
    size = (230,230),
    title = L"$f(x) = -12x_2 + 4x_1^2 + 4x_2^2 + 4x_1x_2$",
    titlefontsize = 8
)

quiver!([0],[0],gradient=([1],[0]), color=:1)
annotate!(1.3, -0.1, text(L"$d_1$",9,:bottom, blue))

quiver!([0],[0],gradient=([-0.5],[1]), color=:1)
annotate!(-0.5, 1.05, text(L"$d_2$",9,:bottom, blue))

savefig("H-orthogonal.pdf")
```

```julia
n=1000

f(x) = (x[1]+1)^2 + (x[2]-2)^2

x1 = range(-50,50,length=n);
x2 = range(-50,50,length=n);

z = [f([x1[i],x2[j]]) for j = 1:n, i = 1:n];

contour(x1, x2, z, 
        levels = [0.2, 1, 2.5, 5, 10, 20, 40],
        xaxis = (L"$x_1$", (-2,2)),
        yaxis = (L"$x_2$", (-1,3)),
        clims = (-5, 15),
        cbar = false,
        aspect_ratio = :equal,
        size = (230, 230),
        legend = true,
        title = L"$f(x) = (x_1+1)^2 + (x_2-2)^2$",
        titlefontsize = 8
)

quiver!([0],[0],gradient=([1],[0]), color=:1)
annotate!(1.05, 0.1, text(L"$d_1$",9,:bottom, blue))

quiver!([0],[0],gradient=([0],[1]), color=:1)
annotate!(-0.1, 1.3, text(L"$d_2$",9,:left, blue))

savefig("orthogonal.pdf")
```

```julia
include("unconstrained_methods.jl")

n = 1000

f(x) = -12x[2] + 4x[1]^2 + 4x[2]^2 + 4(x[1]*x[2]) 

xc = coordinate_descent(f, [0.5,1])

x1 = range(-50,50,length=n);
x2 = range(-50,50,length=n);

# Original function values
#Fabricio(23/09/18): had to add [] between x1[i],x[2] to work on 0.61
z = [f([x1[i],x2[j]]) for j = 1:n, i = 1:n];

contour(x1, x2, z, 
        levels = [-11.5, -9, -5, 2, 10, 20, 40],
        xaxis = (L"$x_1$", (-3,1)),
        yaxis = (L"$x_2$", (0,4)),
        clims = (-20, 30),
        cbar = false,
        aspect_ratio = :equal,
        legend = true,
        size = (300, 300),
        titlefontsize = 10
)

plot!([1/2, -1/2, -1], [1, 1, 2], 
    marker=:circle, 
    label = "Conjugate method")

plot!(xc[1,:],xc[2,:],
    label = "Coord. desc.",
    legend = true,
    line = :dash,
    marker = :x,
    title = L"$f(x) = -12x_2 + 4x_1^2 + 4x_2^2 + 4x_1x_2$")

savefig("coord_conj_H.pdf")
```
