```julia
#Ex. 1 - Lecture 9

using Plots, LaTeXStrings
pyplot(size=(350,350))

n=1000

f(x) = x[1]^2 + x[2]^2 
x1 = range(-50,50,length=n);
x2 = range(-50,50,length=n);


# Original function values
#Fabricio(23/09/18): had to add [] between x1[i],x[2] to work on 0.61
z = [f([x1[i],x2[j]]) for j = 1:n, i = 1:n];

contour(x1, x2, z, 
        levels = [1, 4, 9, 15],
        xaxis = (L"$x_1$", (-4,4)),
        yaxis = (L"$x_2$", (-4,4)),
        clims = (0, 20),
        #clabels = true,    
        title = L"$f(x)=x_1^2 + x_2^2$",
        aspect_ratio = :equal,
        cbar = false
)

plot!(x1, (1 .- x1), color = :green, label = L"$x_1 + x_2 = 1$")        
scatter!([1/2],[1/2], color = :orange, label = L"optimal $x$")

savefig("ex1.pdf")

```

```julia

f(x, μ) = x[1]^2 + x[2]^2 + μ*(x[1] + x[2] -1)^2
x1 = range(-50,50,length=n);
x2 = range(-50,50,length=n);

μ = 0.5

# Original function values
z = [f([x1[i],x2[j]], μ) for j = 1:n, i = 1:n];

contour(x1, x2, z, 
        levels = [1, 4, 9, 15],
        xaxis = (L"$x_1$", (-4,4)),
        yaxis = (L"$x_2$", (-4,4)),
        clims = (0, 20),
        aspect_ratio = :equal,
        title = L"$f_{\mu}(x)=x_1^2 + x_2^2 + 0.5(x_1 + x_2 - 1)^2$",
        cbar = false
)

plot!(x1, (1 .- x1), color = :green, alpha = 0.5, label = "")   

scatter!([1/2],[1/2], color = :orange, alpha = 0.5, label = "")
scatter!([μ/(2μ+1)],[μ/(2μ+1)], color =:green, label = L"optimal $x_\mu$")

savefig("ex2-1.pdf")
```

```julia
μ = 1

z = [f([x1[i],x2[j]], μ) for j = 1:n, i = 1:n];

contour(x1, x2, z, 
        levels = [1, 4, 9, 15],
        xaxis = (L"$x_1$", (-4,4)),
        yaxis = (L"$x_2$", (-4,4)),
        clims = (0, 20),
        aspect_ratio = :equal,
        title = L"$f_{\mu}(x)=x_1^2 + x_2^2 + 1(x_1 + x_2 - 1)^2$",
        cbar = false
)

plot!(x1, (1 .- x1), color = :green, alpha = 0.5, label = "")   

scatter!([1/2],[1/2], color = :orange, alpha = 0.5, label = "")
scatter!([μ/(2μ+1)],[μ/(2μ+1)], color =:green, label = L"optimal $x_\mu$")

savefig("ex2-2.pdf")
```
