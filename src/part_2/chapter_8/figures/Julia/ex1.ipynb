```julia
#Nonlinear optimisation
#Ex. 1 - Lecture 8

# Last edit: 02.09.2021

using Plots, LaTeXStrings
pyplot()

n = 500

x1 = range(0, stop=4, length=n)
plot(x1, (4 .- x1), fill=(10,0.2), legend=false)

x1 = range(-1, stop=10, length=n);
x2 = range(-1, stop=10, length=n);
f(x) = (x[1])^2 + (x[2])^2 

# Original function values
z = [f([x1[i], x2[j]]) for j = 1:n, i = 1:n];

my_levels = [2+i for i = 2:2:20]
contour!(x1,x2,z,linetype = :contour, 
    levels = my_levels, 
    # levels = [4, 8, 12, 16],
    xaxis = (L"$x_1$", (0,4), (0:1:10)),
    yaxis = (L"$x_2$", (0,4)),
    clims = (1,15),
    #   colorbar = true,  
      contour_labels = true,
      aspect_ratio = :equal,
      size=(325,325)  
)

    #size=(1000,700))

#plot!(x1, zeros(500), color = :black)
#plot!(zeros(500), x1, color = :black)
scatter!([2],[2], color = :orange)


savefig("ex1_1.pdf")
```

```julia
u = range(0,40,length=n)
uneg = range(-40,0,length=n)

θpos = -0.5 * (u.^2) + 4 * u
θneg = 4*uneg
    
plot(u, θpos, 
    legend=false,
    xaxis = (L"$u$", (-2,10), (-6:2:10)),
    yaxis = (L"$\theta(u)$", (-5,10), (-6:2:10)),
    aspect_ratio = :equal,
    size=(325,325))

plot!(uneg, θneg)
scatter!([4],[8])
savefig("ex1_2.pdf")
```

```julia
y = range(-10,10,length=n)
vline([0], color=:black, label="")
plot!(y, (4 .- y).^2, fillrange= 0.5*(4 .- y).^2, fill=(0.1), alpha=0, 
    label = L"$G$",
    xaxis = (L"$y$", (-2,4)),
    yaxis = (L"$z$", (0,40), (0:8:40)),
    color = 3,
#    aspect_ratio = :equal,
   size=(400,250))


plot!(y, 0.5*(4 .- y).^2, color= 1, label=L"$v(y)$")

plot!(y, -4*y .+ 8, color = 2, linestyle=:dash, label=L"$z + \overline{u}y = \alpha$")
scatter!([0],[8], color = 2, label = "")

savefig("ex1_3.pdf")
```
