```julia
# Nonlinear optimisation
# Ex. 2 - Lecture 8
# Last edit: 02.09.2021

using Plots, LaTeXStrings
pyplot()

n = 500
x = range(-1,5,length=n)

plot(x, 3 .- x, label = L"$x_1 + x_2 = 3$"),
plot!(x, -3 .+ 2*x, label =L"$-2x_1 + x_2 = -3$")

scatter!([0, 0, 4, 4, 1, 2],[0, 4, 4, 0, 2, 1],
    label = L"$X$",
    xaxis = (L"$x_1$", (-0.1,4.1), (0:1:4)),
    yaxis = (L"$x_2$", (-0.1,4.1), (0:1:4)),
    aspect_ratio = :equal,
    legend = :top,
    size=(325,325)
    )

savefig("ex2_1.pdf")
```
