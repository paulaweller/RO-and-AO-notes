```julia
using Plots, LaTeXStrings
pyplot()

#f(x) = sqrt(abs(x[1]) + abs(x[2]));
f(x) = (x[1]^2 + x[2] - 11)^2 + (x[1] + x[2]^2 - 7)^2 #Himmelblau's function 


n = 200;
x1 = range(-5,5,length=n);
x2 = range(-5,5,length=n);

# Original function values
#Fabricio(23/09/18): had to add [] between x1[i],x[2] to work on 0.61
z = [f([x1[i],x2[j]]) for j = 1:n, i = 1:n];
```

```julia
surface(x1,x2,z,
        xlabel = L"$x_1$",
        ylabel = L"$x_2$",
        title = L"$(x_1^2 + x2 - 11)^2 + (x_1 + x_2^2 - 7)^2$",
        size=(500,500))
#savefig("himmelblau_surface.pdf")
```

```julia
contour(x1,x2,z,  
        xlabel = L"$x_1$",
        ylabel = L"$x_2$",
        title = L"$(x_1^2 + x2 - 11)^2 + (x_1 + x_2^2 - 7)^2$",
        # levels = [collect(0:10:500)],
        size=(500,500))
#savefig("himmelblau_contours.pdf")
```
