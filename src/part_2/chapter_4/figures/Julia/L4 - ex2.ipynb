```julia
#Ex. 2 - Lecture 4

using Plots, LaTeXStrings
pyplot()

n=500
x1 = range(0, stop=sqrt(5)-10e-6, length=n)
x2 = sqrt.((x1.^2).*(-1) .+ 5)

plot(x1, x2, legend=false, fill=(0,0.1), color = :2)
x1 = range(0, stop=3, length=n)

plot!(x1, (3 .- x1), color = :green, fill=(0,0.1))
f(x) = (x[1] - 3)^2 + (x[2] - 2)^2 
x1 = range(-1, stop=10, length=n);
x2 = range(-1, stop=10, length=n);

z = [f([x1[i],x2[j]]) for j = 1:n, i = 1:n];

contour!(x1, x2, z, 
        levels = [0.01, 0.1, 0.5, 1.0, 2.0, 4.0],
        xaxis = (L"$x_1$", (0,4)),
        yaxis = (L"$x_2$", (0,4)),
        size = (325,325),        
        clims = (0,5.5),
        aspect_ratio = :equal,
        clabels = true
)
x1 = [1, 2]
x2 = [2, 1]
λ = 0.2

df(x) = λ*[2(x[1]-3), 2(x[2] - 2)]

df1 = df(x1)
df2 = df(x2)

orange = RGBA(247/255,134/255,21/255,1.0)
light_purple = RGBA(146/255,58/255,123/255, 1.0)



quiver!([x1[1]],[x1[2]], gradient=([df1[1]],[df1[2]]), color = orange)
annotate!([(0.1, 1.8, text(L"$\nabla f(\overline{x})$", 10, :left, orange))])

quiver!([x2[1]],[x2[2]], gradient=([df2[1]],[df2[2]]), color = light_purple)
annotate!([(1.2, 0.4, text(L"$\nabla f(\overline{x})$", 10, :left, light_purple))])

scatter!((1,2), color = orange)
scatter!((2, 1), color = light_purple)


savefig("ex2.pdf")
```
