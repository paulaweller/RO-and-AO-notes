```julia
# Ex. 1 - Lecture 4
using Plots, LaTeXStrings, ForwardDiff
pyplot()

f(x) = (x[1] - 1.5)^2 + (x[2] - 5)^2 

n = 500;
x1 = range(-1,10,length=n);
x2 = range(-1,10,length=n);

# Original function values
z = [f([x1[i],x2[j]]) for j = 1:n, i = 1:n];
```

```julia
# Level curves
dark_purple = RGBA(68/255,42/255,105/255,1.0)

contour(x1,x2,z, 
        levels = [0.05, 1, 4.25, 12, 25, 41, 50],
        xaxis = (L"$x_1$", (0,8)),
        yaxis = (L"$x_2$", (0,8)),
        size = (325,325),
        clims = (0,30),
        fill = false,
        contour_labels = true,
        aspect_ratio = :equal,
        legend = false,
        cbar = false)
        # color = cgrad([dark_blue, blue, light_blue], [0.2, 0.5, 0.9]))

# Constraints
# plot!(Shape([(0,0), (0,2), (1,3), (5.5,0)]), alpha = 0.5, color = 1) # Feasible region
plot!(x1, 2 .+ x1, fill=(true,0.1), color = 2)
plot!(x1, (11 .- 2*x1)/3, color= 3, fill=(true,0.1),xaxis = (L"$x_1$", (0,8)),
        yaxis = (L"$x_2$", (0,8)))

# Dot and gradient of f at dot
x0 = [1,3]
λ = 0.5
df = λ*[2(x0[1]-1.5), 2(x0[2] - 5)]

# Arrow
quiver!([1],[3], gradient=([df[1]],[df[2]]), color = dark_purple)

# Dot
scatter!([x0[1]], [x0[2]], color = dark_purple)

# Arrow label
annotate!([(0.6, 0.7, text(L"$\nabla f(\overline{x})$",10,:left, 
dark_purple))])
# Save figure
savefig("ex1.pdf")
```
