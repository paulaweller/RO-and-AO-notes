```julia
x1 = range(-10, 10, length = 50)
x2 = copy(x1)

# Some convex functions
# f(x,y) = x^2 + 2y^2
#f(x,y) = exp(x)+ exp(y)
# f(x1,x2) = 2x1 + 6x2 - 2x1^2 - 3x2^2 + 4x1*x2


#Quasi convex
# f(x,y) = log(x^2 + y^2)
# f(x,y) = (x^2 + y^2)^(1/2)
f(x,y) = max(x,y)

surface(x1, x2, f, 
    color =:winter,
    xaxis = (-10,10),
    yaxis = (-10,10)
)

```

```julia
contour(x1, x2, f,
    color = :winter,
    xaxis = (-10,10),
    yaxis = (-10,10)
)
```

```julia
f(x) = 2x[1] + 6x[2] - 2x[1]^2 - 3x[2]^2 + 4x[1]*x[2]

d₀ = [2, 3]
x₀ = [1, 2]

θ(λ) = f(x₀ + λ * d₀)

λg = range(-10,10, length = 200)

plot(λg ,θ)
```
